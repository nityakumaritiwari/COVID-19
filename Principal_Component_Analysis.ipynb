```python
#MIN MAX Scaling function
def min_max_scale(X):
 from sklearn.preprocessing import MinMaxScaler
 scaler = MinMaxScaler()
 X = scaler.fit_transform(X)
 return X
```

```python
#Function for Principal component analysis
#95% of variance
def pca_analysis(X):
 from sklearn.decomposition import PCA
 pca = PCA(n_components = 0.95)
 pca.fit(X)
 X = pca.transform(X)
 return X
```

```python
chunksize = 750
tfr = pd.read_csv(path, chunksize=chunksize, iterator=True,header= None)
```

```python
X_array =[]
Y_array= []
Y_DF= pd.DataFrame(Y_array)
X_DF = pd.DataFrame(X_array)
X_DF
```

```python
#Doing min_max and pca on individual chunks of the large csv files 
# and concactnating the result dataframes together
i=1
with tfr as reader:
    for chunk in reader:
        print("index_value")
        print(i)
        X = chunk.iloc[:,1:]
        X = min_max_scale(X)
        X = pca_analysis(X)
        X = pd.DataFrame(X)
        X_DF = pd.concat([X_DF,X], axis=0, ignore_index=True)
        Y = chunk.iloc[:, 0]
        Y_DF = pd.concat([Y_DF,Y], axis=0, ignore_index=True)
        i=i+1
        
```

```python
X_DF
```

```python
#Removing extra NAN values with 0
X_DF = X_DF.fillna(0)

#Joining X and Y
X= X_DF.to_numpy()
Y= Y_DF.to_numpy()
X = np.hstack((Y,X))
df= pd.DataFrame(X)


#Converting first column to int 
#and leaving every other column to float
firstCol = list(df.columns)[0]
df[firstCol] = df[firstCol].astype(int)
df.dtypes
```

```python
#Saving dataframe to csv

df.to_csv("/content/drive/MyDrive/pcaresult.csv", index=False, header=False)
```
